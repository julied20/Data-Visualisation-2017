```python
import pandas as pd
import numpy as np

import os
```

```python
df = pd.read_csv(os.path.join('datasets', 'switzerland_explo.csv'))
```

```python
df = df[['Aggregate Level', 'Commodity Code', 'Commodity','Netweight (kg)', 'Trade Value (US$)']]
```

```python
df2 = df[df['Aggregate Level'] == 4]
df2['Commodity Code'] = df2['Commodity Code'].astype(np.int)
df2 = df2[df2['Commodity Code'] < 2400]
# 2400 = Commodity Code for Tobacco
```
